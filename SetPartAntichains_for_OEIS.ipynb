### 0. The basic code with the class for managing generation of formal concepts via NextClosure algorithm

```python
import numpy as np

class nnBitContext:
    def __init__(self, n=3, i=0, m=3):
            self.attr_names=[str(k+1) for k in range(n)]
            self.objects=[]
            self.obj_names=[str(k+1) for k in range(m)]
            #print(bin(i))
            for j in range(m):                
                self.objects.append(i & ((1<<n) - 1))
                #print(bin(i & ((1<<n) - 1)))
                i=i<<1
                #print(bin(i))
            self.nAttr=len(self.attr_names)
            self.nObj=len(self.objects)
            self.name="Context"

      
 
    def aClosure(self,aSet):
        #вычисление признакового замыкания
        B=(1<<self.nAttr)-1
        for g in self.objects:
            if aSet & g == aSet:
                B=B & g        
        return B
    
    
def cxtFromFile(filename):
    cxt=nnBitContext(0,0,0)
   
    f = open(filename, "r")
    print(f.readline())
    print(f.readline())
        
    cxt.nObj=int(f.readline())
    cxt.nAttr=int(f.readline())
    
    print(f.readline())
    
    for i in range(cxt.nObj):
        cxt.obj_names.append(f.readline())
    
    for i in range(cxt.nAttr):
        cxt.attr_names.append(f.readline())
    

    for l in range(cxt.nObj):
        o=f.readline().replace('.','0').replace('x','1')
        print(o)
        
        cxt.objects.append(int(o,2))
        #print(bin(i & ((1<<n) - 1)))
        
        #print(bin(i)
    cxt.name=filename[:-4]
    f.close()
    return cxt  


    
def BitNextClosure(A,bCxt):
    #print("Bit NextClosure")
    for m in range(bCxt.nAttr-1,-1,-1):
        #print("m=",m)
        if A & (1<<m):
            #print("A=",A)
            A= A & ~(1<<m)
            #print("A= A & ~(1<<m)",A)
        else:
            B=bCxt.aClosure(A |(1<<m))
            #print("A=",A,bin(A))
            #print("B=cl(AUm)",B)
            B_A=B & ~A
            #print("B-A",B_A,bin(B_A))
            if not B_A & ((1<<m)-1):
                #print("B_A", B_A, bin(B_A), "((1<<m)-1)", ((1<<m)-1), bin(((1<<m)-1)))  
                return B
    print("Branch terminated with -1")
    return -1     
 

def FirstClosure(bCxt):
    #print("First closure")
    if bCxt.objects==[]:
        return 2**bCxt.nAttr-1
    eCl=bCxt.objects[0]
    if eCl:
        for g in bCxt.objects[1:]:
            eCl= eCl & g 
  
    return eCl

def CountAllClosures(bCxt):
    
    #print("All closures")
    cnt=0
    prevcnt=0
    A=FirstClosure(bCxt)
    #print("First Closure",A,bin(A))
    while A!=-1:
        #print ("A=",A, "binary representation", bin(A))
        cnt+=1
        A=BitNextClosure(A,bCxt)
        if cnt%1000000==0: 
            if cnt!=prevcnt: 
                print(cnt)
                prevcnt=cnt
    return cnt
```

### 1. Loading formal contexts for antichains and maximal antichains of a partition lattice

```python
AC_cxts=[cxtFromFile("AC_Part"+str(n)+".cxt") for n in range(6)]
```

```python
MA_cxts=[cxtFromFile("MA_Part"+str(n)+".cxt") for n in range(6)]
```

### 2. Generation of maximal antichains for n in {0,1,2,3,4,5} 

```python
%%time
MAs=[CountAllClosures(c) for c in  MA_cxts[:6]]
```
